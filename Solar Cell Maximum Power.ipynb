Using Python, plot the output current and power versus voltage of a solar cell at 25C with the following characteristics: short-circuit current I_sc = 8.3 A and open-circuit voltage V_oc = 0.61 V. Then determine the maximum power output (Pmax) and the corresponding voltage and current at Pmax.


```python

#Ivan Defaz

import numpy as np
import matplotlib.pyplot as plt

# If you want to find a new plot, please change any of the three (3) data below

# Data______________________________________

I_sc = 8.3  # A 
V_oc = 0.61  # V
T_cell = 25 # K

#___________________________________________

T_cello = (T_cell+273)  # K

q = 1.602 * 10 **(-19)  # J/V
n = 1
k = 1.38 * 10 **(-23)   # J/K

I_L = I_sc
I_o = I_L / (np.exp((q * V_oc) / (n * k * T_cello)) - 1)

V_del = 10 **(-3)
V_cell = np.arange(0, V_oc, V_del)

I_cell = I_L - I_o * (np.exp((q * V_cell) / (n * k * T_cello)) - 1)
P_cell = V_cell * I_cell

im = np.argmax(P_cell)
Pmax = P_cell[im]
Vmax = V_cell[im]
Imax = I_cell[im]

plt.plot(V_cell, np.c_[I_cell, P_cell], label=['$I_{cell}$ [A]', '$P_{cell}$ [W]'])
         
plt.legend() 
plt.xlabel('$V_{cell}$ [V]')
plt.grid(True)

print('\nThe Maximum Power Output (Pmax) and the Corresponding Voltage and Current at Pmax')
print( 'Power = ',round(P_cell[im],2),'[W],','Voltage = ', round(V_cell[im],2),'[V], ','Intensity = ',round(I_cell[im],2),'[A]\n')
```
